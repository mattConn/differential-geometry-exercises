```julia
N = 20
u = linspace(0, 2π, N)
v = linspace(0, π, N)
X = u'
Y = sin(u) * sin(v)'
Z = repeat(cos(v)',outer=[N, 1])

#@show supported_args()
# @show supported_types()
# @show supported_styles()
plot(X,Y,Z,linetype=:surface,color="red")
```

```julia
#animation (good)
using Plots
using Calculus
pyplot()

p = 2
f(t) = 1
g(t) = t
n=2
x(t) = f(t)*cos(n)-g(t)*sin(n)
y(t) = f(t)*sin(n)+cos(n)*g(t)
T(t) = sin(t)

COLOR = "black"
color_list = ["black","red"]
LW = 2.0
LABEL = ""

plot()
anim = @animate for i in 0:10
for j in 0:0.1:2pi
    n = j
    plot!(x,y,T,0,i,label=LABEL,color="black")
end
end
gif(anim, "mygif.gif", fps = 5)
# plot!()
```
