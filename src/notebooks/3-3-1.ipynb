```julia
using Plots
using Calculus
pyplot()

p = 2
q = 3

x(t) = p*cos(t)
y(t) = q*sin(t)

COLOR = "black"
LW = 2.0
LABEL = ""
```

```julia
xl(t) = x(pi/4)
yl(t) = t

plot(x, y,0, 2pi,ticks=false,axis=false,label=LABEL,color=COLOR,linewidth=LW)
plot!(xl, yl,y(-pi/4), y(pi/4),ticks=false,axis=false,label=LABEL,color=COLOR,linewidth=LW)

annotate!([(x(pi/4)+.1, y(pi/4)+.2, text("V",22))])
annotate!([(x(-pi/4)+.2, y(-pi/4)-.3, text("W",22))])

scatter!(
    [x(pi/4),x(-pi/4)],[y(pi/4),y(-pi/4)],
    color=COLOR,label=LABEL,marker = (6,6,stroke(0,:black,1.0)))
```
