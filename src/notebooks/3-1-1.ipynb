```julia
using Plots
using Calculus
pyplot()

r(t) = 1+a*cos(t)
x(t) = r(t)*cos(t)
y(t) = r(t)*sin(t)

COLOR = "black"
LW = 2.0
LABEL = ""
```

```julia
a = 0.7
plot(x, y,0, 2pi,ticks=false,axis=false,label=LABEL,color=COLOR,linewidth=LW)
```

```julia
a = 1
plot(x, y,0, 2pi,ticks=false,axis=false,label=LABEL,color=COLOR,linewidth=LW)
```

```julia
a = 2
plot(x, y,0, 2pi,ticks=false,axis=false,label=LABEL,color=COLOR,linewidth=LW)
```
