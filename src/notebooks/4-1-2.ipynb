```julia
using Plots
using Calculus
pyplot()

p = 2
q = 3

x(t) = p*cos(t)
y(t) = q*sin(t)
T(t) = t

COLOR = "black"
LW = 2.0
LABEL = ""
```

```julia
plot(x,y,T,-10,10,ticks=false,axis=false,label=LABEL,color=COLOR,linewidth=LW)
```

```julia
x(t)
surface(x,y,T)
```

```julia
N = 20
u = linspace(0, 2π, N)
v = linspace(0, π, N)
X = u'
Y = sin(u) * sin(v)'
Z = repeat(cos(v)',outer=[N, 1])

#@show supported_args()
# @show supported_types()
# @show supported_styles()
plot(X,Y,Z,linetype=:surface,color="red")
```
